```
import numpy as np 
import pandas as pd
```

```
df = pd.read_csv('USA_cars_datasets.csv', index_col = 0)
df.head()
```

```
#Ver el tipo de datos de las columnas
df.dtypes
```

```
#Ver los colores de todas las columnas
color = df[['color']]
color

#Ver los valores únicos 
np.unique(color)
```

```
#Usando crosstab para ver cuántas veces se repite cada color
freqTable = pd.crosstab(index = df['color'], columns= 'count')
freqTable.sort_values(by = 'count')
```

```
#Usando agrupaciones para ver cuántas veces se repite cada color
freqTable2 = df.groupby('color').count()[['brand']]
freqTable2
```

```
#Usando value_counts para ver cuántas veces se repite cada color
freqTable3 = pd.DataFrame(df['color'].value_counts()).reset_index()
freqTable3.columns = ['color', 'count']
freqTable3
```

```
#Obtener la moda de los colores
df['color'].mode()
```

```
import seaborn as sbs
import matplotlib.pyplot as plt
```

```
sbs.barplot(x = 'color', y = 'count', data = freqTable3)
```

```
barplot = sbs.barplot(x = 'color', y = 'count', data = freqTable3)
barplot = barplot.set_xticklabels(barplot.get_xticklabels(), rotation = 45)
```

```
#Subtabla con colores repetidos más de 10 veces
subTable = freqTable3[freqTable3['count'] > 10]
subTable
```

```
barplot = sbs.barplot(x = 'color', y = 'count', data = subTable)
barplot = barplot.set_xticklabels(barplot.get_xticklabels(), rotation = 45, horizontalalignment = 'right')
```

```
#Diagrama de Pareto 
#A la subtabla le agregamos una columna para obtener la suma acumulada de los valores
subTable['cum_perc'] = subTable['count'].cumsum()

#Para obtener el porcentaje
subTable['cum_perc'] = subTable['count'].cumsum() / subTable['count'].sum() * 100
subTable
```

```
#Empaquetado de datos, una función puede regresar más de un valor
#fig = barras/gráfica 
#ax1 = coordenadas en x y de la suma de valores 
#ax2 = coordenadas en x y del porcentaje 
fig, ax1 = plt.subplots(figsize = (10, 6))
ax1 = sbs.barplot(x = 'color', y = 'count', data = subTable)
ax2 = ax1.twinx() #mismos valores en x que ax1

ax2 = sbs.lineplot(x = 'color', y = 'cum_perc', data = subTable, sort = False)
plt.show()
```
